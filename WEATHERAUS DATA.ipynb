###  Find out the non-null counts of the columns and save the result as a dataframe.

```python
def notnull_data(data):
    
    not_null = data.notnull().sum().sort_values(ascending = False)
    
    return pd.DataFrame(not_null).rename(columns={0:"NotNull"})

notnull_data(df)
```

### Find out the null counts of the columns and save the result as a dataframe.

```python
def isnull_data(data):
    
    isnull = data.isnull().sum().sort_values(ascending = True)
    
    return pd.DataFrame(isnull).rename(columns={0:"Null"})

isnull_data(df)
```

### Find out the unique counts of the columns and save the result as a dataframe

```python
def nunique_data(data):
    
    nunique = data.nunique().sort_values(ascending = False)
    
    return pd.DataFrame(nunique).rename(columns={0:"Nunique"})

nunique_data(df)
```

### Merge the dataframes you created in questions 4-5-6-7. Expected output: example result

```python
def summry_data(data):
    
    df_type = data.dtypes
    
    Not_Null_value =  data.notnull().sum()
    
    Null_value = data.isnull().sum()
    
    percent_value = (data.isnull().sum()/data.isnull().count()*100)
    
    number_unique_value = data.nunique()
    
    df_final = pd.concat([df_type, Not_Null_value, Null_value, percent_value,number_unique_value],
                         axis=1, keys=["Dtypes", 'Notnull', "Isnull" ,'Percent',"Nunique"])
    
    df_final = df_final.sort_values(by="Percent",ascending=True).reset_index().rename(columns={"index":"Columns"})
    
    df_final.Percent = round(df_final.Percent,2)
    
    return df_final

summry_data(df)
```
